```python
"""
Plot a function y = sin^2 x for -2 pi <= x <= 2 pi
Approach:
Calculate and plot 1000 (x, y) points, and store them in the list ax and ay. 
Set up ax list as abcissa - we can't use range directly because that method only produces interger sequences
first wor out the spacing between each x value as 

Delta_x = (xmax - xmin) / (n - 1)
Abcissa points are then 
xi = xmin + iDeltax for i = 0, 1, 2, .... n-1

The corresponding y-axis points are y_i = sin^2(x_i)
"""

import math
import matplotlib.pyplot as plt
xmin, xmax = -2 *math.pi, 2 * math.pi #declaracion de la funcion xmin y xmax 
n = 1000 #Puntos a graficar
x = [0] * n # creacion de una lista en x
y = [0] * n # creacion de una lista en y
dx = (xmax -xmin)/(n-1) # ecuacion a graficar
for i in range (n): #loop for 
    xpt = xmin + i * dx
    x[i] = xpt 
    y[i] = math.sin(xpt)**2 #ecuacion de la ordenada

plt.plot(x, y)
plt.show()
```

# Linspace and Vectorization 
Con el fin de acortar las funciones para graficar, por ejemplo en ejercicio anterior. Se puede emplear la libreria Numpy

The regularly spaced grid of x-coordinates, x, can be crated using linspace. It takes a start value, an end value, and the number of values in the sequence and generates an array (Numpy) of values representing the arithmetic progression between 

Ex: 
x = np.linspace(-5, 5, 1001) - creates the sequence: '.50, -4.99, -4.98, ..... 4.99, 5.00.

Numpy equivalents of math module's methods can act on iterable objects (such as list or Numpy arrays)

Thus, y = np.sin(x) creates a sequence of values (Numpy ndarray), which are sin(xi) for each value xi in the array x


```python
# Vectorization - List and tuples can be turned into array objects. 
import numpy as np
import matplotlib.pyplot as plt
xmin, xmax = -2 *math.pi, 2 * math.pi #declaracion de la funcion xmin y xmax 
n = 1000 #Puntos a graficar
x = np.linspace(xmin, xmax, n)
y = np.sin(x)**2

plt.plot(x, y, 'r')
plt.show()
```

```python
"""
Problem 3.1.2
Michaelis.Menten equation models the kinetics of enzymatic reaction as 

v = d[P]/dt = Vmx[S]/Km + [S]

v = rate of the reaction converting the substrate (S), to product (P) catalyzed by the enzyme. 
Vmax maximum rate (when all the enzyme is bound to S) and the Michaelis constante (Km), is the substrate concentration
at which the re rxt rate is half its maximum value. 

Plot v vs [S] for a rxt with 
Km = 0.04 M 
Vmx = 0.1 Ms-1
"""
#import numpy as np
#import matplotlib.pyplot as plt
# Michaelis.Menten constantes (M), maximum rate (M.s-1)

Km1, Km2, Vmax = 0.04, 0.4, 0.01

#Grid of the sunstrate concentration values (M)

S = np.linspace(0, 2, 1000)

def rate(S, Km, Vmax):
    return Vmax * S / (Km + S)
plt.plot(S, rate(S, Km1, Vmax))
plt.plot(S, rate(S, Km2, Vmax))
plt.show()
```

```python
"""
Problem 3.1.3
The normalized Gaussian function centered at x = 0 is 

g(x) = (1/sigma*sqrt(2pi)exp(-(x**2/2signma**2))

Plot and compare the shapes of these functions for standart dev = 1, 1,5 and 2
"""
#import numpy as np
#import matplotlib.pyplot as plt
x = np.linspace(-6, 6, 1000) 
def gaussian (x, sigma):
    """Return the Normalized Gaussian with StdDev sigma """
    c = np.sqrt(2 * np.pi)
    return np.exp(-0.5 *(x /sigma)**2)/sigma/c
sigma1, sigma2, sigma3 = 1, 1.5, 2
g1, g2, g3 = gaussian(x, sigma1), gaussian(x, sigma2), gaussian(x, sigma3)

plt.plot(x, g1)
plt.plot(x, g2)
plt.plot(x, g3)

plt.show()
```

# Problem 3.2.1 
A molecule, A, reacts to form either B or C with first-order rate constants k1 and k2, respectively. 

d[A]/dt = -(k1 + k2) * [A]

and so 

[A] = [A]0exp^(-k1 + k2)*t

were [A0 is the initical concentration of A 
The pruct concentration (starting form 0) increase in the ratio [B]/[C] = k1/k2 and conservation of matter rqeuires 
[B] + [C] = [A]0 - [A] Therfore 

[B] = (k1 / k1 + k2) * [A]0 * (1 - exp^(-k1 + k2)*t)
[C] = (k1 / k1 + k2) * [A]0 * (1 - exp^(-k1 + k2)*t)

k1 = 300 s-1 
k2 = 100 s-1 
A, B, C 
A0 = 2.0 mol dm-3

```python
#import numpy as np
#import matplotlib.pyplot as plt

k1, k2, = 300, 100 #s-1
A0 = 2. # mol dm-3

#The time axis, in ms 
t = np.linspace(0, 20, 1000)
A = A0 * np.exp(-(k1+k2)*t/1000)
B = (k1/(k1+k2))*(A0-A)
C = (k2/(k1+k2))*(A0-A)

plt.plot(t, A, label = '[A]')
plt.plot(t, B, label = '[B]')
plt.plot(t, C, label = '[C]')
plt.xlabel('time/ms', fontsize = 16)
plt.ylabel('concentration mol/dm^3', fontsize = 16)
plt.title ('Reaction rate plot')
plt.legend()
plt.show()
```

# Multiple Axes 
pyplot.twin() starts a new set of axes with the same x-axis as the original one, but a new y-scale. 
Useful for plotting two or more data series. 

Exemple: 
Correlation over time between the divorce rate in the US state of Maine and the per capita consumption of margarine in that country. Two time series here have different units and meanings and so should be ploted on separate y-axes, sharing a common x-axis (year)

```python
#import matplotlib.pyplot as plt

year = range(2000, 2010)
divorce_rate = [5.0, 4.7, 4.6, 4.4, 4.3, 4.1, 4.2, 4.2, 4.2, 4.1]
margarine_consumption = [8.2, 7.0, 6.5, 5.3, 5.2, 4.0, 4.6, 4.5, 4.2, 3.7]

line1 = plt.plot(year, divorce_rate, 'k-o', label = 'Divorce rate in Maine')
plt.ylabel('Divorce per 1000 people')
plt.legend()

plt.twinx()
line2 = plt.plot(year, margarine_consumption, 'r-s', label = 'Margarine Consumption')
plt.ylabel('lb of Margarine (per capita)')

# Jump through some hoops to get labels in the same legend:
lines = line1 + line2
labels = []

#for line in lines:
    #labels.append(line.gel_label())

plt.legend(lines, labels)
plt.show()
```
